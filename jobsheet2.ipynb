### 4. Operasi matriks dan vektor dengan python

a. Membuat matriks di python

```python
import numpy as np
A = np.array([[80, 30, 25], [50, 20, 10]])
B = np.array([[30, 60, 15], [10, 30, 30]])

C = A + B

print(C)
```

b. Melakukan operasi dot product

```python
import numpy as np
A = np.array([10, 20, 15])
B = np.array([16, 67, 35])

C = np.dot(A, B)

print(C)
```

c. Melakukan operasi cross product

```python
import numpy as np
A = np.array([10, 20, 15])
B = np.array([16, 67, 35])

C = np.cross(A, B)

print(C)
```

### PENUGASAN

PENUGASAN KE 1

```python
import numpy as np
A = np.array([[3, 1],
              [2, 0]])

B = np.array([[2, 1],
              [1, -1]])

C = np.array([[-1, 1, 2],
              [2, 0, -1]])

AB = np.dot(A,B)
BC = np.dot(B,C)

print("Hasil dari A x B adalah = \n", (AB))
print("\n")
print("Hasil dari B . C adalah = \n", (BC))
```

PENUGASAN KE 2

```python
import math
A = 6
B = 8
C = 5

hasilA = math.factorial(A)
hasilB = math.factorial(B)
hasilC = math.factorial(C)

hasilBC = (hasilB // hasilC)

print(hasilA)
print(hasilBC)
```
