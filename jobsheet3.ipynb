# Jobsheet 3 - PENGUKURAN STATISTIK DESKRIPTIF

### 1. Central Tendency

a. Mendapatkan Nilai Mean

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.mean(score)

print(x)
```

b. Mendapatkan Nilai Median

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.median(score)

print(x)
```

c. Mendapatkan Nilai Modus / Mode

```python
from scipy import stats

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = stats.mode(score, keepdims=True)

print(x)
```

### 2. Spread

a. Mendapatkan Nilai Range

```python
import numpy as np

data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]

minimum = np.min(data)
maximum = np.max(data)

x = maximum - minimum

print(x)
```

b. Mendapatkan Nilai Percentile

```python
import numpy as np

data = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.percentile(data, 90)

print(x)
```

c. Mendapatkan Nilai Quartile dan Interquartile Range

```python
import numpy as np

data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1

print(Q1,",", Q3,",", IQR)
```

d. Mendapatkan Nilai Variance dan Standart Deviation

```python
import numpy as np

data = [86, 87, 88, 86, 87, 85, 86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)

print(mean,",", var,",", std)
```

Penugasan ke 3

```python
import numpy as np

data_nilai = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10,
              53, 63, 65, 50, 59, 62, 50, 51, 57, 60,
              63, 65, 65, 53, 99]

data_urut = sorted(data_nilai)

bawah = data_urut[:12]
atas = data_urut[13:]

Q1 = np.median(bawah)
Q3 = np.median(atas)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = []
for x in data_urut:
    if x < lower_bound or x > upper_bound:
        outliers.append(x)

print("a. Urutkan data dari yang terkecil sampai data yang terbesar \n", data_urut, "\n")
print("b. Identifikasi nilai first quartile (Q1) dan third quartile (Q3), dan IQR \n", "Q1 =", Q1, ", Q3 =", Q3, ", IQR =", IQR, "\n")
print("c. Tentukan nilai upper bound dan lower bound \n", "lower bound =", lower_bound, ", upper bound =", upper_bound, "\n")
print("d. Nilai yang menjadi outlier adalah =", outliers)
```

Penugasan ke 4

```python
import numpy as np

berat_badan = np.array([43, 48, 52.5, 58, 63, 68])
frekuensi = np.array([2, 3, 6, 4, 5, 2])

mean = np.sum(berat_badan * frekuensi) / np.sum(frekuensi)
var = np.sum(frekuensi * (berat_badan - mean)**2) / np.sum(frekuensi)
std = np.sqrt(var)

mean = round(mean, 2)
std = round(std, 2)

print("Rata-rata berat badan =", mean)
print("Standar deviasi berat badan =", std)
```
